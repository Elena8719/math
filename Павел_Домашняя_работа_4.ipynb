## 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

$$M(X) = \frac{a + b}{2}$$

$$D(X) = \frac{(b - a)^2}{12}$$

```python
import numpy as np
```

```python
a = 200
b = 800

mean = (a + b) / 2
print(f'Cреднее значение: {mean}')
A = np.array([a, b])
output = np.std(A)
print(f'Дисперсия: {output}')
```
